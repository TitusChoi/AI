## 1. Sigmoid 활성화 함수

활성화 함수 역할을 해주는 sigmoid 활성화 함수의 정의는 다음과 같다.
$$f(x) = {\frac{1}{1 + e^{-x}}}$$
이를 통해 예측 모델을 만들어보자.

```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

x = 1 # 입력값
y = 0 # 기대값
w = tf.random.normal([1], 0, 1) # initial weight
y_p = sigmoid(x * w)            # y predict : 신경망 연산을 수행
print(y_p)
```

```python
a = 0.1 # a는 학습률
for index in range(1000): # 1000번의 학습
    y_p = sigmoid(x * w)
    error = y - y_p
    
    # 학습 또는 갱신
    w = w + x * a * error
    if index % 100 == 0:
        print('반복 횟수 : {0}, 에러값 : {1}, 결과값 : {2}'.format(index, error, y_p))
```

```python
# 조건 변경, 입력값과 출력값 변경

x = 0 # 입력값 1 -> 0
y = 1 # 기대값 0 -> 1
w = tf.random.normal([1], 0, 1) # 초기 가중치
a = 0.1 # a는 학습률
for index in range(1000):
    y_p = sigmoid(x * w)
    error = y - y_p
    
    # 학습 또는 갱신
    w = w + x * a * error
    if index % 100 == 0:
        print('반복 횟수 : {0}, 에러값 : {1}, 결과값 : {2}'.format(index, error, y_p))
```

위 식은 왜 에러값에 따라서 해결하지 못했을까?<br>
입력 값이 0이고, 0이 곱해졌기 때문에 학습이 제대로 수행하지 못한 것이다.<br>
그렇기 때문에 이를 방지하기 위한 아이디어가 필요하다.<br>
이때 등장하는 개념이 바로 편향이다.

```python
# 조건 변경, 입력값과 출력값 변경, 편향 적용

x = 0 # 입력값 1 -> 0
y = 1 # 기대값 0 -> 1
w = tf.random.normal([1], 0, 1) # 초기 가중치
b = tf.random.normal([1], 0, 1) # 초기 편향
a = 0.1 # a는 학습률
for index in range(1000):
    y_p = sigmoid(x * w + 1 * b)
    error = y - y_p
    
    # 학습 또는 갱신
    w = w + x * a * error
    b = b + 1 * a * error

    # 전부 출력하는 것이 아니라 100번 반복마다 출력해보자.
    if index % 100 == 0:
        print('반복 횟수 : {0}, 에러값 : {1}, 결과값 : {2}'.format(index, error, y_p))
```

이러한 경우를 방지하기 위해 편향을 넣는다.<br>
편향($b$)의 입력으로는 보편적으로 1을 곱해서 넣는다.<br>
편향은 $w$처럼 난수 형태로 넣어준다.<br>
따라서 예측값($y_p$)의 최종 식은 다음과 같다.
$$y_p = f(x \times w + 1 \times b)$$
