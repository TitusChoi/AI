## 2. 활성화 함수(Activation Function)
이번에는 신경망에서 활성화 함수 역할을 해주는 sigmoid 함수를 사용해 모델을 학습해보자.<br>
우선, sigmoid 함수의 정의는 다음과 같다.
$$f(x) = {\frac{1}{1 + e^{-x}}}$$
이를 통해 예측 모델을 만들어보자.

```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

x = 1 # 입력값
y = 0 # 기대값
w = tf.random.normal([1], 0, 1) # initial weight
y_p = sigmoid(x * w)            # y predict : 신경망 연산을 수행
print(y_p)
```

```python
# 조건 변경, 입력값과 출력값 변경

x = 0 # 입력값 1 -> 0
y = 1 # 기대값 0 -> 1
w = tf.random.normal([1], 0, 1) # 초기 가중치
lr = 0.1 # a는 학습률
for index in range(1000):
    y_p = sigmoid(x * w)
    error = y - y_p
    
    # 학습 또는 갱신
    w = w + x * lr * error
    if index % 100 == 0:
        print('반복 횟수 : {0}, 에러값 : {1}, 결과값 : {2}'.format(index, error, y_p))
```

위 식은 왜 에러값에 예측을 수행하지 못했을까?<br>
입력 값이 0이고, 0이 곱해졌기 때문에 학습을 제대로 수행하지 못한 것이다.<br>
그렇기 때문에 이를 방지하기 위한 아이디어가 필요하다.<br>
이때 등장하는 개념이 바로 편향($b$)이다.

```python
# 조건 변경, 입력값과 출력값 변경, 편향 적용

x = 0 # 입력값 1 -> 0
y = 1 # 기대값 0 -> 1
w = tf.random.normal([1], 0, 1) # 초기 가중치
b = tf.random.normal([1], 0, 1) # 초기 편향
lr = 0.1 # 학습률
b_x = 1 # bias에 넣어주기 위한 input

epochs = 1000 # epoch 정의 : 학습횟수
for index in range(epochs):
    y_p = sigmoid(x * w + b_x * b)
    error = y - y_p
    
    # 학습 또는 갱신
    w = w + x * lr * error
    b = b + b_x * lr * error

    # 전부 출력하는 것이 아니라 100번 반복마다 출력해보자.
    if index % 100 == 0:
        print('반복 횟수 : {0}, 에러값 : {1}, 결과값 : {2}'.format(index, error, y_p))
```

이러한 경우를 방지하기 위해 편향을 넣는다.<br>
편향($b$)의 입력으로는 보편적으로 1을 곱해서 넣는다.<br>
편향($b$)은 가중치($w$)처럼 난수 형태로 넣어준다.<br>
따라서 예측값($y_p$)의 최종 식은 다음과 같다.
$$y_p = f(x \times w + 1 \times b)$$

## 3. 네트워크(Network)
좀 더 의미있는 연산을 하는 AND 네트워크를 만들어보자.

```python
import numpy as np

# AND 연산자 신경망 구현
x = np.array([[1, 1], [1, 0], [0, 1], [0, 0]])
y = np.array([[1], [0], [0], [0]])

# 파라미터 초기화 과정, mean 0, std dev 1
w = tf.random.normal([2], 0, 1)
b = tf.random.normal([1], 0, 1)
b_x = 1 # bias에 넣어주기 위한 input

epochs = 4000 # 학습횟수
lr = 0.1 # 학습률
length_datasets = len(x)

for i in range(epochs):
    error_sum = 0
    for j in range(length_datasets):
        y_p = sigmoid(np.sum(x[j] * w) + b_x * b)
        error = y[j][0] - y_p
        
        # 학습 또는 갱신
        w = w + x[j] * lr * error
        b = b + b_x * lr * error
        error_sum += error # 누적 에러
        
        # 전부 출력하는 것이 아니라 100번 반복마다 출력해보자.
    if i % 200 == 0:
        print('반복 횟수 : {0}, 에러값 : {1}, 누적에러값 : {2}'.format(i, error, error_sum))
```

```python
# 값이 실제로 잘 예측됐는지 출력
for i in range(length_datasets):
    print("X : ", x[i], ", Y : ", y[i], ", Output : ", sigmoid(np.sum(x[i] * w) + b))
```

위의 결과를 보면 4000 epochs을 취했을 때, AND 연산 값에 근접해가는 네트워크가 성공적으로 구현됐음을 알 수 있다.<br>
이번에는 Y값을 조정해 OR 네트워크를 구현해보자.

```python
# OR 연산자 신경망 구현
x = np.array([[1, 1], [1, 0], [0, 1], [0, 0]])
y = np.array([[1], [1], [1], [0]]) # 하나라도 TRUE이면 TRUE이므로 [1, 0], [0, 1] index 수정

# 파라미터 초기화 과정, mean 0, std dev 1
w = tf.random.normal([2], 0, 1)
b = tf.random.normal([1], 0, 1)
b_x = 1 # bias에 넣어주기 위한 input

epochs = 4000 # 학습횟수
lr = 0.1 # 학습률
length_datasets = len(x)

for i in range(epochs):
    error_sum = 0
    for j in range(length_datasets):
        y_p = sigmoid(np.sum(x[j] * w) + b_x * b)
        error = y[j][0] - y_p
        
        # 학습 또는 갱신
        w = w + x[j] * lr * error
        b = b + b_x * lr * error
        error_sum += error # 누적 에러
        
        # 전부 출력하는 것이 아니라 100번 반복마다 출력해보자.
    if i % 200 == 0:
        print('반복 횟수 : {0}, 에러값 : {1}, 누적에러값 : {2}'.format(i, error, error_sum))
```

```python
# 값이 실제로 잘 예측됐는지 출력
for i in range(length_datasets):
    print("X : ", x[i], ", Y : ", y[i], ", Output : ", sigmoid(np.sum(x[i] * w) + b))
```

예측 값이 AND 네트워크와 마찬가지로 원하는 값에 수렴해간다는 것을 확인해볼 수 있다.<br>
그렇다면 마지막으로 NOR 네트워크를 구성해보자.

```python
# XOR 연산자 신경망 구현
x = np.array([[1, 1], [1, 0], [0, 1], [0, 0]])
y = np.array([[0], [1], [1], [0]]) # XOR : odd input TRUE return

# 파라미터 초기화 과정, mean 0, std dev 1
w = tf.random.normal([2], 0, 1)
b = tf.random.normal([1], 0, 1)
b_x = 1 # bias에 넣어주기 위한 input

epochs = 4000 # 학습횟수
lr = 0.1 # 학습률
length_datasets = len(x)

for i in range(epochs):
    error_sum = 0
    for j in range(length_datasets):
        y_p = sigmoid(np.sum(x[j] * w) + b_x * b)
        error = y[j][0] - y_p
        
        # 학습 또는 갱신
        w = w + x[j] * lr * error
        b = b + b_x * lr * error
        error_sum += error # 누적 에러
        
        # 전부 출력하는 것이 아니라 100번 반복마다 출력해보자.
    if i % 200 == 0:
        print('반복 횟수 : {0}, 에러값 : {1}, 누적에러값 : {2}'.format(i, error, error_sum))
```

```python
# 값이 실제로 잘 예측됐는지 출력
for i in range(length_datasets):
    print("X : ", x[i], ", Y : ", y[i], ", Output : ", sigmoid(np.sum(x[i] * w) + b))
```

결과를 살펴보면 제대로 동작하고 있지 않다는 것을 알 수 있다.<br>
특히, epochs가 200일 때부터 마지막까지 어느 일정 값에 에러값이 수렴하고 있다는 것을 볼 수 있다.<br>
왜 이런 결과가 나왔을까?<br>
가중치를 살펴보니 다음과 같은 결과를 얻을 수 있었다.

```python
print('weight :',w)
print('bias :', b)
```

즉, 가중치 $w$와 편향 $b$ 모두 0에 수렴하고 있다는 사실을 알 수 있다.<br>
이러한 문제 때문에 한 때 인공지능의 겨울이 초래되었고, 이러한 단일 퍼셉트론으로는 XOR 네트워크를 만들지 못한다는 것을 Marvin Minsky, Seymour Papert가 증명해내었다.

## 4. 다중 퍼셉트론
퍼셉트론이 여러 개일 때 XOR 네트워크를 구현할 수 있고, 세 개의 퍼셉트론을 통해 XOR 네트워크를 구현해보겠다.

```python
# XOR 연산자 다중 퍼셉트론을 통한 신경망 구현
import matplotlib.pyplot as plt

x = np.array([[0, 0], [1, 0], [0, 1], [1, 1]])
y = np.array([[0], [1], [1], [0]])

# Dense는 모델에서 사용하는 레이어를 정의하는 명령
# units은 레이어를 구성하는 뉴런의 수
# input_shape는 입력의 차원 수
# activation은 활성화 함수
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=2, activation='sigmoid', input_shape=(2,)),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
])

# SGD(확률적 경사하강법 적용)
# lr은 학습률
# loss는 손실함수(mean square error 적용)
model.compile(optimizer=tf.keras.optimizers.SGD(lr=0.1), loss='mse')
model.summary()

history = model.fit(x, y, epochs=4000, batch_size=1, verbose=0)

# 값이 실제로 잘 예측하는지 출력
print(model.predict(x))

plt.plot(history.history['loss'], 'r--')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.show()
```

따라서 다중 퍼셉트론에서는 XOR 네트워크가 잘 구현된다는 것을 확인할 수 있다.<br>
파라미터는 왜 9개가 나왔을까?<br>
보통 Dense layer의 파라미터 수는 (입력층 뉴런 수)$\times$(출력층 뉴런 수)로 정의된다.<br>
입력층의 경우 (입력 파라미터 수 + 1(편향입력))이라고 볼 수 있다.<br>
따라서 첫 번째 레이어 파라미터 수 (2 + 1) $\times$ 2 = 6이고, 두 번째 레이어 파라미터 수 (2 + 1) $\times$ 1 = 3인 것을 알 수 있다.
