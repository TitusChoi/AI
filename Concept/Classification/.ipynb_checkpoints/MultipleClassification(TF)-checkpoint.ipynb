## 다항 분류(Multiple Classification)

```python
import pandas as pd
import os

# 경로 초기화
os.chdir(r"C:\Users\TitusChoi\Desktop\Library\CodeLion\AI")

red = pd.read_csv('./datasets/winequality-red.csv', sep = ';')
white = pd.read_csv('./datasets/winequality-white.csv', sep = ';')

print(white['quality'].value_counts())
```

```python
# 데이터 간단하게 확인
import matplotlib.pyplot as plt
plt.hist(white['quality'], bins = 7)
plt.show() # 아래와 같은 품질이 있고, 이에 대한 품질 기준을 분류로 만들어보자.
```

```python
red['type'] = 0 # red 비율 0에 할당
white['type'] = 1 # white 비율 1에 할당

wine = pd.concat([red, white])
wine.loc[wine['quality'] < 6, 'new_quality']  = 0
wine.loc[wine['quality'] == 6, 'new_quality']  = 1
wine.loc[wine['quality'] > 6, 'new_quality']  = 2

print(wine['new_quality'].describe())
print(wine['new_quality'].value_counts())
```

```python
# 데이터 정규화 / train, test 데이터 분리
del wine['quality']

# tensorflow를 활용하기 위해 numpy array 활용
import numpy as np
wine_normal = (wine - wine.min()) / (wine.max() - wine.min())
wine_shuffle = wine_normal.sample(frac=1) # 비율에 맞춰서 셔플, 1은 100%
wine_np = wine_shuffle.to_numpy()
```

```python
# 와인 데이터 세트 전체 인덱스 확인
train_idx = int(len(wine_np) * 0.8)
train_X, train_Y = wine_np[:train_idx, :-1], wine_np[:train_idx, -1]
test_X, test_Y = wine_np[train_idx:, :-1], wine_np[train_idx:, -1]

import tensorflow as tf
# 원-핫 인코딩, 학습과 훈련 데이터
train_Y = tf.keras.utils.to_categorical(train_Y, num_classes=3)
test_Y = tf.keras.utils.to_categorical(test_Y, num_classes=3)
```

```python
# tf 분류 구축, 마지막 레이어에 sofrmax activation function 활용
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units = 48, activation='relu', input_shape = (12,)),
    tf.keras.layers.Dense(units = 24, activation='relu'),
    tf.keras.layers.Dense(units = 12, activation='relu'),
    tf.keras.layers.Dense(units = 3, activation='softmax') # 3가지로 분류
])

model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.003), loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
# 와인 분류
history = model.fit(train_X, train_Y, epochs=25, batch_size=32, validation_split=0.25)
```

```python
# 분류 모델 시각화
plt.figure(figsize=(12, 4)) # 시각화 크기 정의

plt.subplot(1, 2, 1) # 시각화 구역 1번에 할당
plt.plot(history.history['loss'], 'b-', label = 'loss')
plt.plot(history.history['val_loss'], 'r--', label = 'val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2) # 시각화 구역 1번에 할당
plt.plot(history.history['accuracy'], 'g-', label = 'accuracy')
plt.plot(history.history['val_accuracy'], 'y--', label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```

```python
# 따로 분류해둔 테스트 값에 모델 평가 시작, 82%라는 것을 알 수 있다.
model.evaluate(test_X, test_Y)
```
