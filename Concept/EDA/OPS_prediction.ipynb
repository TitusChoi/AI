Description   : Exploratory Data Analysis<br>
Type : Time-Series, Regression<br>
Scale : WRMSE(Weighted Root Mean Square Error)<br>
<br>
평가 척도로 활용하는 WRMSE 정의는 아래와 같다.
$$WRMSE = \sum_{i=1}^{N}\sqrt{ \frac{(y_i-\hat{y_i})^2w_i}{\Sigma w_i}}$$
$y_i$는 실제 OPS 값, $\hat{y_i}$는 예측 OPS 값, $w_i$는 가중치이다.
